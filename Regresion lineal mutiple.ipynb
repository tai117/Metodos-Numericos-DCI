```python
import math
import matplotlib.pyplot as plt
import numpy as np
xi = [[1,2,3,1,2,3,1,2,3],
      [0.001,0.001,0.001,0.01,0.01,0.01,0.05,0.05,0.05]]
xi = np.log(xi)
yi = [1.4,8.3,24.2,4.7,28.9,84,11.1,69,200]
yi = np.log(yi)
syi = 0
sx1iyi = 0
sx2iyi = 0
n1 = len(yi)
sx1i = 0
sx2i = 0
sx1i2 = 0
sx1ix2i = 0
sx2i2 = 0
for n in range(0,len(yi)):
  sx1i += xi[0][n]
  sx2i += xi[1][n]
  sx1i2 += xi[0][n]**2
  sx1ix2i += xi[0][n]*xi[1][n]
  sx2i2 += xi[1][n]**2
  syi += yi[n]
  sx1iyi += xi[0][n]*yi[n]
  sx2iyi += xi[1][n]*yi[n]
b = [syi, sx1iyi, sx2iyi]
m = [[n1,sx1i,sx2i],
     [sx1i,sx1i2,sx1ix2i],
     [sx2i,sx1ix2i,sx2i2]]
coef = np.zeros((n,n+1))
#Ingresar los coeficientes del sistema por filas
n = 3
for i in range(0,n):
  for j in range(0,n):
    coef[i][j] = m[i][j]
#Ingresar las soluciones
for i in range(0,n):
  coef[i][n] = b[i]
#Construir la matriz triangular infeior
for k in range(0,n-1):
  for i in range(k+1,n):
    c = coef[i][k]/coef[k][k]
    if coef[k][k] == 0:
      if coef[k+1][k] != 0:
        temp = np.copy(coef[k+1])
        temp2 = np.copy(coef[k])
        coef[k] = coef[k+1]
        coef[k+1] = temp2
    for j in range(k,n+1):
      coef[i][j] = coef[i][j] - c*coef[k][j]
#Normalizar
for i in range(0,n):
  coef[i] = coef[i]/coef[i][i]
#Sustitución hacia atrás
x = np.zeros(n)
x[n-1] = coef[n-1][n]
f = n-2
while f >= 0:
  sum = 0
  g = n-1
  while g > f:
    sum += coef[f][g]*x[g]
    g -= 1
  x[f] = (coef[f][n]-sum)/coef[f][f]
  f -= 1
print('Soluciones')
x[0] = np.exp(x[0])
for i in range(0,n):
  print(f'x{i}','=',x[i])
q = x[0]*(2.5**x[1])*(0.025**x[2])
print('Flujo = ',q)
```

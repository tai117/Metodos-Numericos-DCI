```python
import math
import matplotlib.pyplot as plt
import numpy as np
#Ingresar puntos
xi = [1,2,3,4,5,6,7]
yi = [0.5,2.5,2,4,3.5,6,5.5]
x = np.arange(min(xi)-2,max(xi)+2)
sx2 = 0
sy = 0
sxy = 0
sx = 0 
xprom = sum(xi)/len(xi)
yprom = sum(yi)/len(yi)
sxyimxyp = 0
sximxp2 = 0
syimyp2 = 0
for n in range(0,len(xi)):
  sx2 += xi[n]**2
  sy += yi[n]
  sxy += xi[n]*yi[n]
  sx += xi[n]
  sxyimxyp += (xi[n] - xprom)*(yi[n] - yprom)
  sximxp2 += (xi[n] - xprom)**2
  syimyp2 += (yi[n] - yprom)**2
r = sxyimxyp/np.sqrt(sximxp2*syimyp2)
print ('coeficiente de correlacion = ', r)
a0 = (sx2*sy-sxy*sx)/(len(xi)*sx2-sx**2)
a1 = (len(xi)*sxy-sx*sy)/(len(xi)*sx2-sx**2)
print (a1,'x + ',a0)
plt.scatter(xi,yi,color='m')
plt.plot(x,a1*x+a0,'k')
```

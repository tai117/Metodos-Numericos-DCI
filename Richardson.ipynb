```python
from sympy.utilities import lambdify
import sympy as sp
import numpy as np
import math
import matplotlib.pyplot as plt
x = sp.symbols('x')
f = str(input('Ingrese la función: '))
h = float(input('Ingrese h: '))
h1 = 0.25
x0 = float(input('x0: '))
f = lambdify(x,f)
fp1 = (f(x + h)-f(x - h))/(2*h)
fp1 = lambdify(x,fp1)
fp2 = (fp1(x + 0.5*h1)-fp1(x - 0.5*h1))/(2*h1)
fp2 = lambdify(x,fp2)
d = 4*fp2(x0)/3 - fp1(x0)/3
print(fp1(x0),fp2(x0),d)
```

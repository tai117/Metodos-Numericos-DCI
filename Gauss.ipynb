```python
import numpy as np
n = int(input('Ingrese el tamaño del sistema: '))
coef = np.zeros((n,n+1))
#Ingresar los coeficientes del sistema por filas
for i in range(0,n):
  for j in range(0,n):
    coef[i][j] = float(input(f'Ingrese el coeficiente [{i},{j}]: '))
#Ingresar las soluciones
for i in range(0,n):
  coef[i][n] = float(input(f'Ingrese la solución {i}: '))
#Construir la matriz triangular infeior
for k in range(0,n-1):
  for i in range(k+1,n):
    c = coef[i][k]/coef[k][k]
    for j in range(k,n+1):
      coef[i][j] = coef[i][j] - c*coef[k][j]
#Normalizar
for i in range(0,n):
  coef[i] = coef[i]/coef[i][i]
#Sustitución hacia atrás
x = np.zeros(n)
x[n-1] = coef[n-1][n]
f = n-2
while f >= 0:
  sum = 0
  g = n-1
  while g > f:
    sum += coef[f][g]*x[g]
    g -= 1
  x[f] = (coef[f][n]-sum)/coef[f][f]
  f -= 1
print('Soluciones')
for i in range(0,n):
  print(f'x{i}','=',x[i])
```

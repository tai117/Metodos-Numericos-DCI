```python
import math
import matplotlib.pyplot as plt
import numpy as np
#Ingresar puntos
xi1 = [0.5,1,2,3,4]
yi1 = [10.4,5.8,3.3,2.4,2]
xi = 1/np.sqrt(xi1)
yi = np.sqrt(yi1)
x = np.arange(-1,5,0.1)
sx2 = 0
sy = 0
sxy = 0
sx = 0 
for n in range(0,len(xi)):
  sx2 += xi[n]**2
  sy += yi[n]
  sxy += xi[n]*yi[n]
  sx += xi[n]
a0 = (sx2*sy-sxy*sx)/(len(xi)*sx2-sx**2)
a1 = (len(xi)*sxy-sx*sy)/(len(xi)*sx2-sx**2)
print (a1,'x + ',a0)
plt.scatter(xi1,yi1,color='m')
a = a1/a0
b = 1/a0
plt.plot(x,(((a+np.sqrt(x)))/(b*np.sqrt(x)))**2,'k')
print(((a+np.sqrt(1.6))/(b*np.sqrt(1.6)))**2)
```

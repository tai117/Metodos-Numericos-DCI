```python
import math
import matplotlib.pyplot as plt
import numpy as np
#Ingresar puntos
xi1 = [0.5,0.8,1.5,2.5,4]
yi1 = [1.1,2.4,5.3,7.6,8.9]
yi = 1./np.asarray(yi1)
xi = 1./(np.asarray(xi1)**2)
x = np.arange(0,5,0.1)
sx2 = 0
sy = 0
sxy = 0
sx = 0 
for n in range(0,len(xi)):
  sx2 += xi[n]**2
  sy += yi[n]
  sxy += xi[n]*yi[n]
  sx += xi[n]
a0 = (sx2*sy-sxy*sx)/(len(xi)*sx2-sx**2)
a1 = (len(xi)*sxy-sx*sy)/(len(xi)*sx2-sx**2)
print (a1,'x + ',a0)
plt.scatter(xi1,yi1,color='m')
cs = a1/a0
km = 1/a0
plt.plot(x,(km*x**2)/(cs+x**2),'k')
print ('cs =',cs,'kmax =',km)
```

```python
import math
import matplotlib.pyplot as plt
import numpy as np
#Ingresar puntos
xi = [1, 1.25, 1.5, 1.75, 2]
yi1 = [5.1, 5.79, 6.53, 7.45, 8.46]
yi = np.log(yi1)
x = np.arange(min(xi)-2,max(xi)+2,0.1)
sx2 = 0
sy = 0
sxy = 0
sx = 0 
for n in range(0,len(xi)):
  sx2 += xi[n]**2
  sy += yi[n]
  sxy += xi[n]*yi[n]
  sx += xi[n]
a0 = (sx2*sy-sxy*sx)/(len(xi)*sx2-sx**2)
a1 = (len(xi)*sxy-sx*sy)/(len(xi)*sx2-sx**2)
print (a1,'x + ',a0)
plt.scatter(xi,yi1,color='m')
plt.plot(x,np.exp(a0)*np.exp(a1*x),'k')
```

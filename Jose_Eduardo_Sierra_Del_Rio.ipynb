Desarrolle una ecuación lineal múltiple para pronosticar el área como función de la estatura y el peso. Utilícela para estimar el área superficial de una persona de 187 cm y 78 kg.

```python
def EliminacionGaussiana(a,b):
    n = len(b)
    for k in range(0,(n-1)): 
        if a[k,k] == 0:
            if a[k+1,k] != 0:
                temporal_a = np.copy(a[k,:])
                temporal_b = np.copy(b[k])
                a[k,:]  = a[k+1,:]
                b[k]  = b[k+1]
                a[k+1,:] = temporal_a
                b[k+1] = temporal_b

        for i in range(k+1,n):  
            m = a[i,k]/a[k,k]
            for j in range(0,n):  
                a[i,j] = a[i,j] - m*a[k,j] 
            b[i] = b[i] - m * b[k]

    x = np.zeros(n)
    x[n-1] = b[n-1]/a[n-1,n-1]
    for k in range(n-2,-1,-1):
        s = 0
        for j in range(k,n): 
            s = s + a[k,j] * x[j]
        x[k] = (b[k] - s)/a[k,k]
    return x
```
